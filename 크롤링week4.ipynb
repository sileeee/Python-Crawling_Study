### ë°ì´í„° ìˆ˜ì§‘ ë‚´ìš© csvë¡œ ì €ì¥í•˜ê¸°

```
import requests
from bs4 import BeautifulSoup

# navertv.csvíŒŒì¼ì„ ì“°ê¸°ëª¨ë“œ(w)ë¡œ ì—´ì–´ì¤ë‹ˆë‹¤.
f = open("navertv.csv", "w")
# í—¤ë” ì¶”ê°€í•˜ê¸°
f.write("ì œëª©,ì±„ë„ëª…,ì¬ìƒ ìˆ˜,ì¢‹ì•„ìš” ìˆ˜")

raw = requests.get("https://tv.naver.com/r")
html = BeautifulSoup(raw.text, "html.parser")

# 1ìœ„ - 100ìœ„ ì»¨í…Œì´ë„ˆ ì„ íƒì: dl.cds_info
clips = html.select("dl.cds_info")

for cl in clips:
    # ìˆ˜ì •ëœ ë¶€ë¶„
    title = cl.select_one("dt.title").text.strip()
    chn = cl.select_one("dd.chn").text.strip()
    hit = cl.select_one("span.hit").text.strip()
    like = cl.select_one("span.like").text.strip()

    # replace í•¨ìˆ˜ë¥¼ í™œìš©í•˜ì—¬ ,ë¥¼ ì œê±°í•©ë‹ˆë‹¤.
    title = title.replace(",", "")
    chn = chn.replace(",", "")
    hit = hit.replace(",", "")
    like = like.replace(",", "")

    # replaceí•¨ìˆ˜ë¥¼ í™œìš©í•´ì„œ "ì¬ìƒ ìˆ˜" ì œê±°
    hit = hit.replace("ì¬ìƒ ìˆ˜", "")
    # ìŠ¬ë¼ì´ì‹±ì„ í™œìš©í•´ì„œ "ì¢‹ì•„ìš” ìˆ˜" ì œê±°
    like = like[5:]

    print("ì œëª©", title)
    print("ì±„ë„ëª…", chn)
    print(hit)
    print(like)
    print("="*50)
    
    # íŒŒì¼ì— ë‚´ìš©ì„ ì…ë ¥í•©ë‹ˆë‹¤.    
    f.write(title + "," + chn + "," + hit + "," + like + "\n")

# ì‘ì—…ì´ ëë‚œ íŒŒì¼ì„ ë‹«ìŠµë‹ˆë‹¤.
# ë°˜ë³µë¬¸ ë°–ì—ì„œ ë‹«ì•„ì¤ë‹ˆë‹¤.
f.close()
```
